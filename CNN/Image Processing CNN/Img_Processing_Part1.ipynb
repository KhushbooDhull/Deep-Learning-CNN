### Reading and Image

<b>cv2.imread(img_path,flags)</b>

<b>flags</b><br>
<b>1) cv2.IMREAD_COLOR</b> - It specifies to load a color image. Any transparency of image will be neglected. It is the default flag. Alternatively, we can pass integer value 1 for this flag.<br>
<b>2) cv2.IMREAD_GRAYSCALE</b> - It specifies to load an image in grayscale mode. Alternatively, we can pass integer value 0 for this flag.<br>

### Converting Color Space
1) There can be different color spaces like RGB, HSV (Hue Saturation Value) etc<br>
2) cv2.cvtColor() is used to change color spaces<br>

```python
d1 = cv2.imread('scenary1.png')
d1_rgb = cv2.cvtColor(d1,cv2.COLOR_BGR2RGB)
d1_gray = cv2.cvtColor(d1,cv2.COLOR_BGR2GRAY)
d1_hsv = cv2.cvtColor(d1,cv2.COLOR_BGR2HSV)

plt.imshow(d1),plt.title('Original image in BGR'),plt.show()
plt.imshow(d1_rgb),plt.title('Img in RGB'),plt.show()
plt.imshow(d1_gray,'gray'),plt.title('Img in GRAY'),plt.show()
plt.imshow(d1_hsv),plt.title('Img in HSV'),plt.show()
plt.show()

# cv2.imshow('Original Img',d1)
# cv2.waitKey(0)

# cv2.imshow('Img in RGB',d1_rgb)
# cv2.waitKey(0)

# cv2.imshow('Img in gray',d1_gray)
# cv2.waitKey(0)

# cv2.imshow('Img in hsv',d1_hsv)
# cv2.waitKey(0)

# cv2.destroyAllWindows()
```

<b>cv2.circle(image, center_coordinates, radius, color, thickness)</b> - <br>
thickness - It is thickness of the rectangle border line in px. Thickness of -1 px will fill the rectangle shape by the specified color.

<b>cv2.line(img, start_point, end_point, color, thickness)</b><br>

#### Creating Image from numpy array

```python
g = np.zeros((300,300,3))
g[10:50,20:70] = [255,0,0]  # Red
g[10:50,230:280] = [0,255,0]  # Green
g[170:180,100:200] = [0,0,255]  # Blue
g[100:125,140:160] = [128,128,0]  # Yellow
plt.imshow(g)
plt.show()
```

```python
g1 = np.zeros((300,300,3))
g1[:,:,:] = [255,255,255]
plt.imshow(g1)
plt.show()
```

```python

img = np.zeros((400,512,3))
# Draw a staright green line with thickness of 5 px
img = cv2.line(img,(0,80),(512,80),(0,255,0),5)

# Draw an empty blue rectangle with thickness of 5 px
img = cv2.rectangle(img,(70,115),(170,160),(0,0,255),5)

# Draw a filled rectangle blue rectangle
img = cv2.rectangle(img,(270,115),(370,160),(0,0,255),-1)

# Draw an empty red circle with thickness of 5 px
img = cv2.circle(img,(120,250), 40, (255,80,102), 5)

# Draw an filled red circle
img = cv2.circle(img,(325,250), 40, (255,150,153), -1)

# Draw a staright green line with thickness of 5 px
img = cv2.line(img,(240,0),(240,512),(0,255,0),5)

# Drawing an empty polygon
pts = np.array([[80,325],[95,380],[120,385],[135,330]]) 
# top left, bottem left, bottom right, top right
img = cv2.polylines(img,[pts],True,(0,255,255),3)

# Drawing an filled polygon
pts = np.array([[260,355],[325,383],[360,360],[270,330]])
img = cv2.polylines(img,[pts],True,(0,255,255),30)

font = cv2.FONT_HERSHEY_DUPLEX
img = cv2.putText(img,'Hollow',(20,70), font, 2,(255,255,255),3,cv2.LINE_AA)

font = cv2.FONT_HERSHEY_TRIPLEX
img = cv2.putText(img,'Filled',(280,70), font, 2,(255,255,255),2,cv2.LINE_AA)

plt.imshow(img)
plt.show()
```

### Color Chaneel Split

```python
b = cv2.imread('rgb_channel.png')
b = cv2.cvtColor(b,cv2.COLOR_BGR2RGB)
plt.imshow(b)
plt.show()
```

```python
cc_red,cc_green,cc_blue = cv2.split(b)
plt.imshow(cc_red,'gray'),plt.title('Red Channel'),plt.show()
plt.imshow(cc_green,'gray'),plt.title('Green Channel'),plt.show()
plt.imshow(cc_blue,'gray'),plt.title('Blue Channel'),plt.show()
plt.show()
```
